# Data Modeling 
## Objectif : Entrainer differents modéls de machine learning afin de trouver le mieux qui peut nous permettre de predir s'il y aura  divorce ou non 

```python
#import relevant libraries
from sklearn.model_selection import train_test_split, GridSearchCV
#features extraction
y = df_selec["Class"]
x = df_selec.drop("Class", axis = 1)
x_train,x_test,y_train,y_test = train_test_split(x, y ,test_size=0.40,random_state=0)  #splitting data with test size of 40%
```

# -- K-NN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score,f1_score
knn = KNeighborsClassifier (n_neighbors= 6)
knn.fit (x_train , y_train)
print(knn.score(x_train,y_train))
knn_pred=knn.predict(x_test)   # evaluating our model
print("score:{}".format(accuracy_score(y_test, knn_pred)))
print("score:{}".format(precision_score(y_test, knn_pred)))
print("f1_score:{}".format(f1_score(y_test, knn_pred)))
```

```python
# confusion matrix
confusion_matrix = pd.crosstab(y_test, knn_pred, rownames=['y_test'], colnames=['y_predi'])
sns.heatmap(confusion_matrix, annot=True)
```

# --Decision Tree

```python
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
#applying tree algorithm
param_grid = { 
    
    'max_features': range(1,15,1),
    'max_depth' :  range(1,30,1),
    'criterion' :['gini', 'entropy']
}
grid = GridSearchCV( estimator = DecisionTreeClassifier(), param_grid = param_grid, cv=5)
grid.fit(x_train, y_train)

```

```python
grid.best_params_
```

```python
dectree = DecisionTreeClassifier(criterion= 'gini', max_depth =  7, max_features = 6)
dectree.fit (x_train, y_train)
dt_pred=dectree.predict(x_test)   # evaluating our model
```

```python
print("score:{}".format(accuracy_score(y_test, dt_pred)))
print("score:{}".format(precision_score(y_test, dt_pred)))
print(dectree.score(x_test,y_test))
print("f1_score:{}".format(f1_score(y_test, dt_pred)))

```
